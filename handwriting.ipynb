```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\NASEEMA NASRIN\Desktop\my intership\handwritten_text.csv")
text = " ".join(df["text"].astype(str))

print(text[:300])
```

```python

# Create character vocabulary
chars = sorted(list(set(text)))
char_to_idx = {c: i for i, c in enumerate(chars)}
idx_to_char = {i: c for i, c in enumerate(chars)}
```

```python

# Convert text to integers
encoded = np.array([char_to_idx[c] for c in text])
```

```python
import numpy as np

seq_length = 20

X, y = [], []

for i in range(len(encoded) - seq_length):
    X.append(encoded[i:i + seq_length])
    y.append(encoded[i + seq_length])

X = np.array(X)
y = np.array(y)

print("X type:", type(X))
print("y type:", type(y))
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(chars), 64),
    tf.keras.layers.LSTM(128, return_sequences=False),
    tf.keras.layers.Dense(len(chars), activation='softmax')
])

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam'
)
```

```python
model.fit(
    X,
    y,
    epochs=20,
    batch_size=64
)

```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(chars), 128),
    tf.keras.layers.LSTM(256, return_sequences=True),
    tf.keras.layers.LSTM(256),
    tf.keras.layers.Dense(len(chars), activation='softmax')
])
```

```python
def generate_text(model, start_string, length=300):
    input_eval = [char_to_idx[s] for s in start_string]
    input_eval = tf.expand_dims(input_eval, 0)

    generated = []

    for _ in range(length):
        predictions = model(input_eval)
        predicted_id = tf.random.categorical(predictions, 1)[0][0].numpy()

        input_eval = tf.expand_dims([predicted_id], 0)
        generated.append(idx_to_char[predicted_id])

    return start_string + ''.join(generated)

print(generate_text(model, "hello how are you "))
```
